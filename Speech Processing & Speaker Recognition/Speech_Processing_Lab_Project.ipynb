#Introduction à la reconnaissance du locuteur par réseaux de neurones

La reconnaissance du locuteur vise à identifier une personne à partir des caractéristiques uniques de sa voix. Contrairement à la reconnaissance de la parole qui décode le contenu ("que dit-on?"), cette technologie détermine l'identité ("qui parle?") en s'appuyant sur les spécificités acoustiques propres à chaque individu, résultant de facteurs physiologiques et comportementaux.

Les avancées récentes en apprentissage profond ont transformé ce domaine. Les réseaux de neurones surpassent désormais les approches classiques (GMM, SVM) grâce à leur capacité à extraire automatiquement des caractéristiques discriminantes complexes et à modéliser efficacement les variations intra et inter-locuteurs.

Notre projet vise à développer un système capable de reconnaître spécifiquement nos voix parmi d'autres locuteurs, ce qui soulève plusieurs défis techniques: extraction de caractéristiques robustes, conception d'une architecture neuronale adaptée, entraînement efficace avec des données limitées, et développement d'un système temps réel. Ces défis illustrent la complexité et l'intérêt de la reconnaissance du locuteur dans le contexte des interfaces homme-machine modernes.

#Extraction des caractéristiques

```python
import pandas as pd
import numpy as np
import librosa
import io
import soundfile as sf
from tqdm import tqdm
from sklearn.preprocessing import LabelEncoder
```

```python
def load_audio_from_dataframe(csv_path):
    # Charger le DataFrame
    df = pd.read_csv(csv_path)

    audio_data = []
    labels = []

    for _, row in tqdm(df.iterrows(), total=len(df), desc="Chargement des données audio"):
        try:
            # Extraire le dictionnaire (qui est stocké comme chaîne de caractères)
            audio_dict = eval(row['All Audios'])

            # Extraire les bytes audio
            audio_bytes = audio_dict['bytes']

            # Convertir les bytes en signal audio
            # Utiliser BytesIO pour créer un fichier-like object
            with io.BytesIO(audio_bytes) as buffer:
                y, sr = sf.read(buffer)

                # Convertir en mono si stéréo
                if len(y.shape) > 1:
                    y = np.mean(y, axis=1)

            # Ajouter le signal audio et son label à nos listes
            audio_data.append(y)
            labels.append(row['Ids'])

        except Exception as e:
            print(f"Erreur lors du chargement de l'entrée {_}: {e}")

    return audio_data, labels, sr
```

Utilisation de la méthode MFCC pour extraire les caractéristiques

Avantages :

1. Modélisent bien la perception humaine de la parole
2. Dimensionnalité réduite (généralement 13-40 coefficients)
3. Efficaces pour capturer les caractéristiques du conduit vocal

```python
def extract_mfcc(signal, sample_rate, n_mfcc=13, n_fft=2048, hop_length=512, include_deltas=True):

    # Extraction des MFCC
    mfccs = librosa.feature.mfcc(y=signal,
                                 sr=sample_rate,
                                 n_mfcc=n_mfcc,
                                 n_fft=n_fft,
                                 hop_length=hop_length
                                 )

    # Normalisation des MFCC
    mfccs = librosa.util.normalize(mfccs, axis=1)

    if include_deltas:
        # Calcul des deltas (première dérivée)
        delta_mfccs = librosa.feature.delta(mfccs)

        # Calcul des delta-delta (seconde dérivée)
        delta2_mfccs = librosa.feature.delta(mfccs, order=2)

        # Concaténation des caractéristiques
        features = np.concatenate([mfccs, delta_mfccs, delta2_mfccs])
        return features
    else:
        return mfccs
```

La fonction `pad_or_truncate` assure que toutes les caractéristiques audio (comme les MFCC) ont exactement la même longueur avant d'être traitées par le réseau de neurones. Elle fonctionne selon deux scénarios:

Si l'échantillon est trop court (moins de frames que `max_len`):

- Elle ajoute des zéros à la fin (padding)
- Cela préserve toutes les informations originales
- C'est comme ajouter du silence après la parole


Si l'échantillon est trop long (plus de frames que `max_len`):

- Elle coupe l'excédent (troncature)
- On garde uniquement le début de l'enregistrement
-Cela suppose que le début contient suffisamment d'informations pour identifier le locuteur

```python
def pad_or_truncate(features, max_len):

    if features.shape[1] < max_len:
        # ajouter des zéros à droite
        pad_width = max_len - features.shape[1]
        features_padded = np.pad(features, ((0, 0), (0, pad_width)), mode='constant')
    else:
        # couper pour correspondre à max_len
        features_padded = features[:, :max_len]
    return features_padded
```

```python
csv_path = "/content/drive/MyDrive/TP/Project/vox_dataset.csv"

# Charger les données audio et les labels
audio_data, labels, sample_rate = load_audio_from_dataframe(csv_path)
print(f"Nombre de fichiers audio chargés: {len(audio_data)}")
```

```python
import matplotlib.pyplot as plt

mfcc_lengths = []

for audio in tqdm(audio_data, desc="Calcul des longueurs MFCC"):
    mfcc_features = extract_mfcc(audio, sample_rate)
    mfcc_lengths.append(mfcc_features.shape[1])  # nombre de frames

plt.figure(figsize=(10,6))
plt.hist(mfcc_lengths, bins=50, color='skyblue', edgecolor='black')
plt.title("Distribution des longueurs de MFCC (n_frames)")
plt.xlabel("Nombre de frames")
plt.ylabel("Nombre d'échantillons")
plt.grid(True)
plt.show()

print(f"Min: {np.min(mfcc_lengths)} frames")
print(f"Max: {np.max(mfcc_lengths)} frames")
print(f"Moyenne: {np.mean(mfcc_lengths):.2f} frames")
print(f"Médiane: {np.median(mfcc_lengths)} frames")
print(f"90ème percentile: {np.percentile(mfcc_lengths, 90)} frames")
```

```python
# Extraire les MFCC pour chaque signal audio
features_list = []
fixed_length = 424

for audio in tqdm(audio_data, desc="Extraction des MFCC"):
    mfcc_features = extract_mfcc(audio, sample_rate)
    mfcc_padded = pad_or_truncate(mfcc_features, fixed_length)
    features_list.append(mfcc_padded)

# Convertir en arrays numpy
X = np.array(features_list)

# Encodage des labels
label_encoder = LabelEncoder()
y = label_encoder.fit_transform(labels)

print(f"Forme des caractéristiques extraites: {X.shape}")
print(f"Classes: {label_encoder.classes_}")
```

#Développement du modèle


```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, precision_score, recall_score, f1_score
import seaborn as sns
import time
```

Cette classe permet de charger efficacement les données dans PyTorch:

* convertit les caractéristiques `MFCC` en tenseurs PyTorch
* ajoute une dimension pour le canal
* facilite le chargement par lots (batching) pendant l'entraînement

```python
class SpeakerDataset(Dataset):
    def __init__(self, features, labels):
        self.features = features
        self.labels = labels

    def __len__(self):
        return len(self.labels)

    def __getitem__(self, idx):
        # Ajout d'une dimension pour le canal (équivalent à une image avec 1 canal)
        feature = self.features[idx]
        feature = np.expand_dims(feature, axis=0)  # Forme: [1, n_features, sequence_length]
        return torch.FloatTensor(feature), torch.LongTensor([self.labels[idx]])[0]
```

Le modèle utilise une architecture `CNN` à plusieurs niveaux:

1. Blocs convolutionnels (4 blocs):

* Chaque bloc contient une couche convolutionnelle 2D, une normalisation par lot, une activation ReLU et un pooling max
* Les couches convolutionnelles extraient des caractéristiques spatiales de plus en plus abstraites des spectrogrammes MFCC
* Le nombre de filtres augmente progressivement (16→32→64→128) pour capter des motifs de plus en plus complexes
* Les opérations de pooling réduisent la taille spatiale pour diminuer la complexité et augmenter la robustesse


2. Couches fully connected:

* Les caractéristiques extraites sont aplaties puis passées dans un réseau dense
* Les couches successives (256→128→num_classes) réduisent progressivement la dimension
* Des couches de dropout (0.5 et 0.3) sont ajoutées pour prévenir le surapprentissage

```python
# Définition de l'architecture du modèle CNN
class SpeakerRecognitionCNN(nn.Module):
    def __init__(self, num_classes):
        super(SpeakerRecognitionCNN, self).__init__()

        # Premier bloc convolutionnel
        self.conv1 = nn.Sequential(
            nn.Conv2d(1, 16, kernel_size=(3, 3), stride=1, padding=1),
            nn.BatchNorm2d(16),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=(2, 2))
        )

        # Deuxième bloc convolutionnel
        self.conv2 = nn.Sequential(
            nn.Conv2d(16, 32, kernel_size=(3, 3), stride=1, padding=1),
            nn.BatchNorm2d(32),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=(2, 2))
        )

        # Troisième bloc convolutionnel
        self.conv3 = nn.Sequential(
            nn.Conv2d(32, 64, kernel_size=(3, 3), stride=1, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=(2, 2))
        )

        # Quatrième bloc convolutionnel
        self.conv4 = nn.Sequential(
            nn.Conv2d(64, 128, kernel_size=(3, 3), stride=1, padding=1),
            nn.BatchNorm2d(128),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=(2, 2))
        )

        # Calcul dynamique de la taille après les convolutions
        # Pour MFCC avec deltas (39, 424) après 4 couches de MaxPool2d(2,2)
        # La nouvelle taille sera (39/16, 424/16) = (2, 26)
        self.fc_input_size = 128 * 2 * 26

        # Couches fully connected
        self.fc = nn.Sequential(
            nn.Linear(self.fc_input_size, 256),
            nn.ReLU(),
            nn.Dropout(0.5),
            nn.Linear(256, 128),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(128, num_classes)
        )

    def forward(self, x):
        # Appliquer les blocs convolutionnels
        x = self.conv1(x)
        x = self.conv2(x)
        x = self.conv3(x)
        x = self.conv4(x)

        # Aplatir pour les couches fully connected
        x = x.view(x.size(0), -1)

        # Appliquer les couches fully connected
        x = self.fc(x)
        return x
```

Cette fonction gère l'ensemble du processus d'entraînement:

1. Elle utilise une boucle d'époques pour entraîner le modèle
2. À chaque époque, elle:

* Effectue une passe avant (`forward pass`) sur les données d'entraînement
* Calcule la perte avec la fonction de perte croisée-entropie
* Effectue une rétropropagation (`backward pass`) pour calculer les gradients
* Met à jour les poids du modèle avec l'optimiseur Adam
* Évalue le modèle sur l'ensemble de validation


3. Elle implémente l'arrêt précoce (`early stopping`) pour éviter le surapprentissage
4. Elle sauvegarde le meilleur modèle selon la perte de validation

```python
# Fonction d'entraînement
def train_model(model, train_loader, val_loader, criterion, optimizer, num_epochs, device, patience=10):
    model.to(device)
    best_val_loss = float('inf')
    patience_counter = 0
    train_losses = []
    val_losses = []
    train_accuracies = []
    val_accuracies = []

    for epoch in range(num_epochs):
        # Mode entraînement
        model.train()
        train_loss = 0.0
        train_correct = 0
        train_total = 0

        for inputs, labels in tqdm(train_loader, desc=f"Epoch {epoch+1}/{num_epochs} [Train]"):
            inputs, labels = inputs.to(device), labels.to(device)

            # Réinitialiser les gradients
            optimizer.zero_grad()

            # Forward pass
            outputs = model(inputs)
            loss = criterion(outputs, labels)

            # Backward pass et optimisation
            loss.backward()
            optimizer.step()

            # Statistiques
            train_loss += loss.item() * inputs.size(0)
            _, predicted = torch.max(outputs.data, 1)
            train_total += labels.size(0)
            train_correct += (predicted == labels).sum().item()

        # Calculer la perte et la précision moyennes pour l'époque
        epoch_train_loss = train_loss / len(train_loader.dataset)
        epoch_train_acc = train_correct / train_total
        train_losses.append(epoch_train_loss)
        train_accuracies.append(epoch_train_acc)

        # Mode évaluation
        model.eval()
        val_loss = 0.0
        val_correct = 0
        val_total = 0

        with torch.no_grad():
            for inputs, labels in tqdm(val_loader, desc=f"Epoch {epoch+1}/{num_epochs} [Val]"):
                inputs, labels = inputs.to(device), labels.to(device)

                # Forward pass
                outputs = model(inputs)
                loss = criterion(outputs, labels)

                # Statistiques
                val_loss += loss.item() * inputs.size(0)
                _, predicted = torch.max(outputs.data, 1)
                val_total += labels.size(0)
                val_correct += (predicted == labels).sum().item()

        # Calculer la perte et la précision moyennes pour l'époque
        epoch_val_loss = val_loss / len(val_loader.dataset)
        epoch_val_acc = val_correct / val_total
        val_losses.append(epoch_val_loss)
        val_accuracies.append(epoch_val_acc)

        print(f"Epoch {epoch+1}/{num_epochs} - "
              f"Train Loss: {epoch_train_loss:.4f}, Train Acc: {epoch_train_acc:.4f} - "
              f"Val Loss: {epoch_val_loss:.4f}, Val Acc: {epoch_val_acc:.4f}")

        # Early stopping
        if epoch_val_loss < best_val_loss:
            best_val_loss = epoch_val_loss
            patience_counter = 0
            # Sauvegarder le meilleur modèle
            torch.save(model.state_dict(), '/content/drive/MyDrive/TP/Project/Model/best_speaker_model.pth')
            print(f"Meilleur modèle sauvegardé avec une perte de validation de {best_val_loss:.4f}")
        else:
            patience_counter += 1
            if patience_counter >= patience:
                print(f"Early stopping à l'époque {epoch+1}")
                break

    # Charger le meilleur modèle
    model.load_state_dict(torch.load('/content/drive/MyDrive/TP/Project/Model/best_speaker_model.pth'))

    return model, train_losses, val_losses, train_accuracies, val_accuracies
```

Cette fonction évalue le modèle entraîné sur l'ensemble de test:

1. Elle calcule plusieurs métriques de performance:

* ***Précision (accuracy)***: proportion de prédictions correctes
* ***Precision***: proportion de prédictions positives qui sont correctes
* ***Rappel (recall)***: proportion de vrais positifs qui sont correctement identifiés
* ***F1-score***: moyenne harmonique de la précision et du rappel


2.  Elle mesure le temps d'inférence pour évaluer la rapidité du modèle
3.  Elle génère et affiche une matrice de confusion
4. Elle produit un rapport de classification détaillé

```python
# Fonction d'évaluation
def evaluate_model(model, test_loader, device, class_names):
    model.to(device)
    model.eval()

    all_predictions = []
    all_labels = []

    # Mesurer le temps d'inférence
    inference_times = []

    with torch.no_grad():
        for inputs, labels in tqdm(test_loader, desc="Évaluation"):
            inputs, labels = inputs.to(device), labels.to(device)

            # Mesurer le temps d'inférence
            start_time = time.time()
            outputs = model(inputs)
            inference_time = time.time() - start_time
            inference_times.append(inference_time)

            _, predicted = torch.max(outputs.data, 1)

            all_predictions.extend(predicted.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())

    # Calculer les métriques
    accuracy = accuracy_score(all_labels, all_predictions)
    precision = precision_score(all_labels, all_predictions, average='weighted')
    recall = recall_score(all_labels, all_predictions, average='weighted')
    f1 = f1_score(all_labels, all_predictions, average='weighted')
    conf_matrix = confusion_matrix(all_labels, all_predictions)
    avg_inference_time = sum(inference_times) / len(inference_times)

    # Afficher les résultats
    print(f"Précision (Accuracy): {accuracy:.4f}")
    print(f"Precision: {precision:.4f}")
    print(f"Rappel (Recall): {recall:.4f}")
    print(f"F1-Score: {f1:.4f}")
    print(f"Temps d'inférence moyen par échantillon: {avg_inference_time*1000:.2f} ms")

    # Afficher la matrice de confusion
    plt.figure(figsize=(10, 8))
    sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues',
                xticklabels=class_names, yticklabels=class_names)
    plt.xlabel('Prédictions')
    plt.ylabel('Vraies étiquettes')
    plt.title('Matrice de confusion')
    plt.tight_layout()
    plt.savefig('/content/drive/MyDrive/TP/Project/confusion_matrix.png')
    plt.show()

    # Afficher le rapport de classification détaillé
    print("\nRapport de classification:")
    print(classification_report(all_labels, all_predictions, target_names=class_names))

    return accuracy, precision, recall, f1, avg_inference_time
```

Cette fonction orchestre l'ensemble du processus:

1. Elle configure les `hyperparamètres`:

* ***Taille de batch:*** 32 (bon équilibre entre vitesse et mémoire)
* ***Taux d'apprentissage:*** 0.001 (valeur standard pour Adam)
* ***Nombre d'époques:*** 50 (avec early stopping)
* ***Régularisation L2:*** 1e-5 (pour réduire le surapprentissage)


2. Elle crée les dataloaders pour charger efficacement les données
3. Elle initialise le modèle, définit la fonction de perte et l'optimiseur
4. Elle lance l'entraînement et visualise les courbes d'apprentissage
5. Elle évalue le modèle final sur l'ensemble de test
6. Elle sauvegarde le modèle et les résultats pour une utilisation ultérieure

```python
# Fonction principale
def main():
    # Configuration du device
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    print(f"Utilisation de: {device}")

    # Hyperparamètres
    batch_size = 32
    learning_rate = 0.001
    num_epochs = 50
    weight_decay = 1e-5

    # Créer les datasets
    train_dataset = SpeakerDataset(X_train, y_train)
    val_dataset = SpeakerDataset(X_val, y_val)
    test_dataset = SpeakerDataset(X_test, y_test)

    # Créer les dataloaders
    train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
    val_loader = DataLoader(val_dataset, batch_size=batch_size)
    test_loader = DataLoader(test_dataset, batch_size=batch_size)

    # Initialiser le modèle
    num_classes = len(np.unique(y_train))
    model = SpeakerRecognitionCNN(num_classes)

    # Définir la fonction de perte et l'optimiseur
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.Adam(model.parameters(), lr=learning_rate, weight_decay=weight_decay)

    # Entraîner le modèle
    model, train_losses, val_losses, train_accuracies, val_accuracies = train_model(
        model, train_loader, val_loader, criterion, optimizer, num_epochs, device
    )

    # Visualiser les courbes d'apprentissage
    plt.figure(figsize=(12, 5))

    # Graphique de la perte
    plt.subplot(1, 2, 1)
    plt.plot(train_losses, label='Perte d\'entraînement')
    plt.plot(val_losses, label='Perte de validation')
    plt.xlabel('Époque')
    plt.ylabel('Perte')
    plt.title('Courbes de perte')
    plt.legend()

    # Graphique de la précision
    plt.subplot(1, 2, 2)
    plt.plot(train_accuracies, label='Précision d\'entraînement')
    plt.plot(val_accuracies, label='Précision de validation')
    plt.xlabel('Époque')
    plt.ylabel('Précision')
    plt.title('Courbes de précision')
    plt.legend()

    plt.tight_layout()
    plt.savefig('/content/drive/MyDrive/TP/Project/learning_curves.png')
    plt.show()

    # Évaluer le modèle
    class_names = label_encoder.classes_
    accuracy, precision, recall, f1, inference_time = evaluate_model(
        model, test_loader, device, class_names
    )

    # Enregistrer les résultats
    results = {
        'accuracy': accuracy,
        'precision': precision,
        'recall': recall,
        'f1': f1,
        'inference_time': inference_time,
        'batch_size': batch_size,
        'learning_rate': learning_rate,
        'weight_decay': weight_decay,
    }

    print("Résultats finaux:")
    for key, value in results.items():
        print(f"{key}: {value}")

    # Sauvegarder le modèle final
    torch.save({
        'model_state_dict': model.state_dict(),
        'optimizer_state_dict': optimizer.state_dict(),
        'class_names': class_names,
        'results': results,
    }, '/content/drive/MyDrive/TP/Project/Model/speaker_recognition_model.pth')

    return model, results
```

```python
# Paramètres
SR = 16000  # Fréquence d'échantillonnage
MAX_LEN = 424 # Longueur maximale des features

def upload_audio():
    """Télécharge un fichier audio depuis l'ordinateur local"""
    print("Veuillez télécharger un fichier audio (.wav, .mp3, etc.):")
    uploaded = files.upload()

    # Récupérer le nom du premier fichier téléchargé
    filename = list(uploaded.keys())[0]
    print(f"Fichier téléchargé: {filename}")

    # Convertir au bon format et taux d'échantillonnage
    y, sr = librosa.load(filename, sr=SR)

    # Sauvegarder en format WAV avec le bon taux d'échantillonnage
    temp_file = "temp_recording.wav"
    write(temp_file, SR, y)

    print("Audio traité et converti à", SR, "Hz")
    display(Audio(temp_file))

    return temp_file

def extract_features(file_path):
    """Extrait les caractéristiques d'un fichier audio en utilisant extract_mfcc et pad_or_truncate"""
    # Chargement du fichier audio
    y, sr = librosa.load(file_path, sr=SR)

    # Utiliser la fonction extract_mfcc
    features = extract_mfcc(y, sr, n_mfcc=13, include_deltas=True)

    # Normalisation (si pas déjà fait dans extract_mfcc)
    features = (features - np.mean(features)) / (np.std(features) + 1e-8)

    # Ajuster la longueur avec pad_or_truncate
    features = pad_or_truncate(features, MAX_LEN)

    # Préparation pour le modèle
    features = np.expand_dims(features, axis=0)  # Ajouter dimension batch
    features = np.expand_dims(features, axis=0)  # Ajouter dimension canal

    return torch.FloatTensor(features)

def identify_speaker(model_path='/content/drive/MyDrive/TP/Project/Model/speaker_recognition_model.pth'):
    """Identifie le locuteur à partir d'un fichier audio téléchargé"""
    # Télécharger l'audio
    audio_file = upload_audio()

    # Extraire les caractéristiques
    features = extract_features(audio_file)

    # Résoudre le problème de chargement du modèle en désactivant weights_only
    try:
        # Version 1: Utiliser weights_only=False (moins sécurisé mais plus compatible)
        checkpoint = torch.load(model_path, map_location=torch.device('cpu'), weights_only=False)
    except Exception as e:
        print(f"Erreur lors du chargement avec weights_only=False: {e}")

        # Version 2: Ajouter des globals sécurisés (plus sécurisé)
        try:
            # Importer les modules nécessaires pour ajouter des globals sécurisés
            from torch.serialization import add_safe_globals
            import numpy._core.multiarray as ma

            # Ajouter les fonctions numpy nécessaires aux globals sécurisés
            add_safe_globals([ma._reconstruct])

            # Réessayer le chargement
            checkpoint = torch.load(model_path, map_location=torch.device('cpu'))
        except Exception as e2:
            print(f"Erreur lors du chargement avec globals sécurisés: {e2}")
            print("\nSolution alternative: essayons de créer un modèle de test pour démonstration")

            # Créer un modèle de test pour la démonstration
            model = SpeakerRecognitionCNN(2)  # 2 classes pour démonstration
            class_names = ["GMI", "Unknown"]

            # Faire une prédiction aléatoire
            with torch.no_grad():
                outputs = model(features)
                _, predicted = torch.max(outputs, 1)
                predicted_class = class_names[predicted.item()]

            print(f"Démonstration avec un modèle factice - Prédiction: {predicted_class}")
            return predicted_class

    # Récupérer les noms de classes
    class_names = checkpoint['class_names']

    # Initialiser le modèle
    model = SpeakerRecognitionCNN(len(class_names))
    model.load_state_dict(checkpoint['model_state_dict'])
    model.eval()

    # Prédiction
    with torch.no_grad():
        outputs = model(features)
        _, predicted = torch.max(outputs, 1)
        predicted_class = class_names[predicted.item()]

    # Afficher le résultat
    if predicted_class == "GMI":
        print("C'est GOURINE Mohammed Islem")
    else:
        print("Une voix inconnue")

    return predicted_class
```
